```python
import mysql.connector, os, pandas
from dotenv import load_dotenv

load_dotenv()
mysql_password = os.getenv('mysql_password')

cnx = mysql.connector.connect(user ='root', password=mysql_password, host='localhost', database='kolkata_weather')
cursorObject = cnx.cursor()

dataframe= pandas.DataFrame(index= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
query = '''select min(`Minimum Temperature`) from weather group by month(`Date Time`);'''
cursorObject.execute(query)
result = cursorObject.fetchall()
result = round(temp[0],2) for temp in result
result.append(min(result))
dataframe['Minimum Temperature']= [result]

query = '''select avg(`Temperature`) from weather group by month(`Date Time`);'''
cursorObject.execute(query)
result = cursorObject.fetchall()
dataframe['Average Temperature']= [round(temp[0],2) for temp in result]

query = '''select max(`Maximum Temperature`) from weather group by month(`Date Time`);'''
cursorObject.execute(query)
result = cursorObject.fetchall()
dataframe['Maximum Temperature']= [round(temp[0],2) for temp in result]

query = '''select avg(`Precipitation`) from weather group by month(`Date Time`);'''
cursorObject.execute(query)
result = cursorObject.fetchall()
dataframe['Precipitation']= [round(temp[0],2) for temp in result]

query = '''select count(*)/6 from weather where Precipitation>0 group by Month(`Date time`);'''
cursorObject.execute(query)
result = cursorObject.fetchall()
dataframe['Days of Rainfall']= [round(float(temp[0]),2) for temp in result]

dataframe
```
